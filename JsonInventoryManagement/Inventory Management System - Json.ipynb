```python
#importing libraries
import json, time

#reading the available inventory details from the records.json
file = open("Records.json", "r")
data = file.read()
file.close()

#loading the json data into python object - dict from string
records = json.loads(data)

#vairable to store the transaction
sale = ''

#displaying the available products details
print("---------------- MENU ----------------------")
print("Id :","NAME     ", "PRICE", "QTY")
for key in records.keys():
    print(key ,":", records[key]['Name'], "|", "₹",records[key]['Price'], "|",records[key]['Qty'])

#taking user and order details
print("\n------------ DETAILS --------------\n")
ui_name    = str(input("Enter you name       :"))
ui_email   = str(input("Enter your email     :"))
ui_phoneno = str(input("Enter your phone no  :"))
ui_pid     = str(input("Enter the product id   :"))
ui_qty     = int(input("Enter the quantity     :"))

#if we have the available quantity 
if(records[ui_pid]['Qty']>= ui_qty):
    #updating the qty in the records
    records[ui_pid]['Qty'] = records[ui_pid]['Qty'] - ui_qty

    #printing the bill
    print("\n--------------- BILL ------------------------\n")

    print("Product Name     :", records[ui_pid]['Name'])
    print("Product Price    :", "₹",records[ui_pid]['Price'])
    print("Product Qty      :", ui_qty)
    
    #calculating total billing amount
    print("\n-------------------------------------------\n")
    print("Total Amount     :", "₹",(records[ui_pid]['Price']*ui_qty))
    print("\n-------------------------------------------\n")
    
    #storing purchase/transaction details
    sale = ui_name + " | " + ui_email + " | " + ui_phoneno + " | " + records[ui_pid]['Name'] + " | " + "Rs."+ str(records[ui_pid]['Price']) + " | " +  str(ui_qty) +  " | " + "Rs."+ str(records[ui_pid]['Price']*ui_qty) + " | " + time.ctime() + "|" + "\n"

    print("\n Thank You for your Order!")
    print("\n-------------------------------------------\n")

#if we don't have the required quantity available
else:
    print("\n-------------------------------------------\n")
    print("Sorry, we don't have enough quantity.")
    
    #asking the customer if he wants the currently available quantity
    choice = str(input(f"Do you want {records[ui_pid]['Qty']} units ?(Y/N)"))
    
    #if customer agrees to take the available quantity
    if(choice == 'Y' or choice=='y') :
        #printing bill
        print("\n--------------- BILL ------------------------\n")
        print("Product Name     :", records[ui_pid]['Name'])
        print("Product Price    :", "₹",records[ui_pid]['Price'])
        print("Product Qty      :", records[ui_pid]['Qty'])

        print("\n-------------------------------------------\n")
        print("Total Amount     :", "₹",(records[ui_pid]['Price']*records[ui_pid]['Qty']))
        print("\n-------------------------------------------\n")
        
        #storing the purchase/transaction details
        sale = ui_name + " | " + ui_email + " | " + ui_phoneno + " | " + records[ui_pid]['Name'] + " | " + "Rs."+ str(records[ui_pid]['Price']) + " | " +  str(records[ui_pid]['Qty']) +  " | " + "Rs."+ str(records[ui_pid]['Price']*records[ui_pid]['Qty']) + " | " + time.ctime() + "|" + "\n"

        print("\n Thank You for your Order!")
        print("\n-------------------------------------------\n")
        records[ui_pid]['Qty'] = 0
        
    #if customer denies that they don't want the available quantity    
    else:
        print("\n Thank You for your Visit, Hope to see you soon again!")
        print("\n-------------------------------------------\n")
    
#Adding transaction details to the sales.txt
file = open("Sales.txt", "a")
file.write(sale)
file.close()

#Updating the inventory
js = json.dumps(records)
fd = open("Records.json", "w")
fd.write(js)
fd.close()

```
